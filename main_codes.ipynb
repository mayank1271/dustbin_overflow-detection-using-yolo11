```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os

# 沒ｸ Just provide image path (no need to extract filename)
input_image_path = "/content/GvJO_CHXEAAgdEy.jpeg"

# 洫 Load YOLO model
model = YOLO("/content/drive/MyDrive/newdata.1model/weights/best.pt")

# 沐 Predict and save
results = model.predict(source=input_image_path, save=True, conf=0.70, iou=0.5, agnostic_nms=True)

# 沒 Get saved directory
save_dir = results[0].save_dir

# 沒 Auto-detect the output image file (any jpg inside save_dir)
output_image_path = None
for file in os.listdir(save_dir):
    if file.lower().endswith(".jpg"):
        output_image_path = os.path.join(save_dir, file)
        break

if output_image_path is None:
    raise FileNotFoundError("笶 Output image not found in saved directory!")

# 泱ｼｸ Show the output image
img = cv2.imread(output_image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.axis("off")
plt.title("沒ｸ YOLOv11 Prediction Output")
plt.show()
```

```python
# 沐ｧ Install ultralytics
!pip install ultralytics --quiet

from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os

# 笨 Set path of uploaded model and video
# 沁ｯ Make sure to upload these files using Colab upload panel (left sidebar)
model_path = "/content/drive/MyDrive/newdata.1model/weights/best.pt"  # replace with your uploaded model file name
video_path = "/content/twitter_1751997087369.mp4"  # replace with your uploaded video file name

# 笨 Load YOLOv11 model
model = YOLO(model_path)

# 沁･ Load video
cap = cv2.VideoCapture(video_path)

# 汳ｾ Setup video writer to save output
save_output = True
output_path = "/content/yolo_output_video.mp4"
if save_output:
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = cap.get(cv2.CAP_PROP_FPS)
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# 沐 Process video
frame_count = 0
max_frames_to_show = 3

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 沐 Run YOLOv11 detection
    results = model.predict(source=frame, conf=0.60, verbose=False)
    annotated_frame = results[0].plot()

    # 泱ｼｸ Show a few frames
    if frame_count < max_frames_to_show:
        plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
        plt.title(f"YOLOv11 - Frame {frame_count + 1}")
        plt.axis('off')
        plt.show()

    if save_output:
        out.write(annotated_frame)

    frame_count += 1

cap.release()
if save_output:
    out.release()

print("笨 Detection done! Download video below 汨")

# 沒･ Download the saved output
from google.colab import files
files.download(output_path)
```

```python
import cv2
from ultralytics import YOLO

model = YOLO(r"C:\Users\mayank manjhi\Downloads\newdata.1model-20250709T134401Z-1-001\newdata.1model\weights\best.pt")
video_url = "http://192.168.0.104:4747/video"
cap = cv2.VideoCapture(video_url)

while True:
    ret, frame = cap.read()
    if not ret:
        print("笶 Frame not received")
        break

    cv2.imwrite("debug_frame.jpg", frame)  # Save frame to verify visually

    frame = cv2.resize(frame, (640, 640))
    results = model.predict(source=frame, conf=0.80,iou=0.50, verbose=False)

    # Debug: check if detections exist
    print(results[0].boxes)

    if results[0].boxes:
        annotated = results[0].plot()
    else:
        annotated = frame

    cv2.imshow("沒ｱ YOLO with DroidCam", annotated)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
